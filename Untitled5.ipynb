blotting sin linea para trazar solo los marcadores , pueden ser notaciones de cadenas de acceso directo "o" que significan "anillos"
Ejemplo: dibujha dos puntos en el diagrama uno en la pocicion (1,3) y otro en la posicion (8,10)

```python
import matplotlib.pyplot as plt
import numpy as np
xpuntos = np.array ([1,8])
ypuntos = np.array ([3,10])
plt.plot(xpuntos,ypuntos)
plt.show()
```

programa de divicion (scater) se emplean para mostrar numeracion numericas o agrupaciones ejemplo:

```python
import matplotlib.pyplot as plt

altura = [150,160,170,180,190,200]
pesos = [20,60,70,80,90,100]

plt.scatter(altura,pesos,color ="red", marker = "x")
plt.title("altura vs pesos")
plt.xlabel("altura (cm)")
plt.ylabel("peso (kg)")

plt.show()
```

graficos de barras sirven para comparar categorias como variables tanto en orientacion vertical como horizontal
EJEMPLO :de barras verticales

```python
import matplotlib.pyplot as plt
categorias = ["a","b","c","d"]
valores =[4,7,1,8]
plt.bar(categorias,valores,color ="orange")
plt.title("grafica de barras")
plt.xlabel("categorias")
plt.ylabel("valores")
plt.show()
```
